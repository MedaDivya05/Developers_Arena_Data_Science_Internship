```python
import pandas as pd

df = pd.read_csv("sales_data.csv")

print("Dataset Loaded Successfully ")
df.head()
```

```python
print("Rows and Columns:", df.shape)
print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
print("Missing Values:\n")
print(df.isnull().sum())
```

```python
# Remove duplicates if any
df = df.drop_duplicates()

print("Duplicates removed if present")
```

```python
# Total Revenue
total_sales = df["Total_Sales"].sum()

# Best Selling Product
best_product = df.groupby("Product")["Total_Sales"].sum().idxmax()

# Highest Revenue Region
best_region = df.groupby("Region")["Total_Sales"].sum().idxmax()

# Average Sales
avg_sales = df["Total_Sales"].mean()

print("Total Revenue:", total_sales)
print("Best Selling Product:", best_product)
print("Highest Revenue Region:", best_region)
print("Average Sale:", avg_sales)
```

```python
print("="*50)
print("SALES DATA ANALYSIS REPORT")
print("="*50)

print(f"Total Revenue: {total_sales:,.2f}")
print(f"Best Selling Product: {best_product}")
print(f"Highest Revenue Region: {best_region}")
print(f"Average Sales Value: {avg_sales:,.2f}")

print("="*50)
```

```python
product_sales = df.groupby("Product")["Total_Sales"].sum()
product_sales.sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

product_sales.plot(kind="bar")
plt.title("Product Wise Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```
